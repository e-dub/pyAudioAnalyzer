# Guitar - Note A2 (110 Hz)

In this notebook we will go through a basic analysis of audio. First we import the necessary Python library

```python
import pyAudioAnalyzer as paa
```

Next we initialize an instance of the sound analyzer and set the audio file to be analyzed. 

```python
AA = paa.AudioAnalyzer()
AA.read("audioFiles/A-Saite.wav")
```

The audio file of interest is now played. You should hear an "impact test" in which I hit a wine glass with a knife.

```python
AA.play()
```

```python
AA.PlotTimeDomain()
```

```python
AA.PlotTimeDomain(tMin=1.1, tMax=8)
```

After finding the proper range by playing with the values, you can cut the data as follows.

```python
AA.cutData(tMin=1.1, tMax=8)
```

We replot the time domain without giving the minimum and maximum time to check the correctness of the cut. Please note that the time will be rezeroed to your minimum time. 

```python
AA.PlotTimeDomain()
```

We can now choose two strains (engineering) inbetween which the Young's modulus will be calculated. 

```python
AA.PlotFFT()
```

```python
AA.PlotFFT(fMax=1000)
```

```python
AA.PlotFFT(fMin=100, fMax=120)
```

```python
AA.PlotSpectrogram()
```

```python
AA.PlotSpectrogram(fMax=500)
```
